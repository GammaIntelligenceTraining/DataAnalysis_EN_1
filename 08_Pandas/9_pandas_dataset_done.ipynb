### 7. What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?

```python
min_workers = df[df['hours-per-week'] == df['hours-per-week'].min()]
```

```python
round(min_workers[min_workers.salary == '>50K'].shape[0] / len(min_workers) * 100, 1)
```

### 8. People from which native country have the highest percentage of those who earn >50K and what is that percentage?

```python
h_salary = df[df.salary == '>50K']
```

```python
h_salary['native-country'].value_counts()
```

```python
df['native-country'].value_counts()
```

```python
earnings_table = h_salary['native-country'].value_counts() / df['native-country'].value_counts()
```

```python
earnings_table.sort_values(ascending=False).head(1).round(3) * 100
```
