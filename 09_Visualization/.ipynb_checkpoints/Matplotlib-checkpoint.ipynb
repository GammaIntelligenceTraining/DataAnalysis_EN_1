```python
import numpy as np
from matplotlib import pyplot as plt
```

# Create and adjust a linear plot

```python
x = np.arange(3, 9)
y = x ** 2
x2 = np.array([1, 2, 3])
y2 = x2 * 10 + 3
```

```python
plt.plot(x, y, 'g--H')
plt.plot(x2, y2, color='hotpink', linewidth=4, linestyle='-.', marker='|',
         markersize=15, markeredgewidth=5, alpha=.5)
```

# Setp

```python
lines = plt.plot(x, y, x2, y2)
plt.setp(lines, c='#231dc2', lw=3, marker='<', ms=8,
         mfc='yellow', ls='--', alpha=.6, gapcolor='red')
plt.show()
```

# Subplots / Figure layout

```python
plt.subplot(1, 3, 1)
plt.plot(np.random.rand(10))
plt.grid()
plt.subplot(1, 3, 2)
plt.plot(np.random.rand(10))
plt.subplot(1, 3, 3)
plt.plot(np.random.rand(10))
plt.tight_layout()
plt.grid()
```

```python
ax1 = plt.subplot(3, 1, 1)
ax2 = plt.subplot(3, 1, 2)
ax3 = plt.subplot(3, 1, 3)
plt.tight_layout()
ax2.plot(np.random.rand(10))
ax1.grid()
```

## Subplots / object-oriented

```python
fig, ax = plt.subplots(2, 2, sharex=True, sharey=True, width_ratios=[.5, 1], height_ratios=[1.5, 1])
```

```python
fig.set_size_inches(12, 7)
fig.set_facecolor('#9e3ccf')
fig.set_edgecolor('orange')
fig.set_linewidth(5)

ax[0, 1].plot(np.sin(np.arange(-np.pi, np.pi, 0.1)), 'xr')
ax[0, 1].grid()
ax[1, 0].plot(np.random.rand(50), ':y>')

fig
```

## Gridspec

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(3, 3)
mid = (0.5, 0.5)
ax1 = fig.add_subplot(gs[0, :]).annotate('[0, :]', mid)
ax2 = fig.add_subplot(gs[1:, 0]).annotate('[1:, 0]', mid)
ax3 = fig.add_subplot(gs[1:, 2]).annotate('[1:, 2]', mid)
```

# Grids and limits

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.tan(np.arange(-10, 10, 0.1)), 'b--')
plt.xlim(-50, 250)
ax.set_ylim(-100, 100)
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.tan(np.arange(-10, 10, 0.1)), 'b--')
ax.axis((50, 150, -70, 70))
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.tan(np.arange(-10, 10, 0.1)), 'b--')
plt.xticks(np.arange(0, 201, 50))
ax.set_yticks(np.arange(-80, 81, 10))
plt.grid()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.tan(np.arange(-10, 10, 0.1)), 'b--')
plt.xlim(-50, 250)
ax.set_ylim(-100, 100)
ax.minorticks_on()
ax.grid(which='major', lw=3, c='#d4d65c', axis='y')
ax.grid(which='minor', c='g', alpha=.4, ls=':', axis='x')
```

# Titles

```python
font1 = {'family': 'comic sans ms', 'color': 'brown', 'size': 15, 'weight': 'bold'}
font2 = {'family': 'times new roman', 'color': '#4d3369', 'size': 12, 'style': 'italic'}

fig, ax = plt.subplots()
ax.plot(np.cos(np.arange(-np.pi, np.pi, 0.2)), '-.o')
plt.title('THIS IS A TITLE!', loc='right', fontdict=font1)
plt.suptitle('SUPERTITLE!')
ax.set_xlabel('X AXIS', fontdict=font2)
plt.ylabel('Y AXIS', loc='bottom', fontdict=font2)
```

# Legend

```python
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot()
ax.plot(np.sin(np.linspace(-np.pi/2, np.pi/2, 20)), 'cx:', label='Sine wave')
ax.plot(np.cos(np.linspace(-np.pi/2, np.pi/2, 20)), '--r8', label='Cosine wave')
ax.legend()
```

```python
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot()
ax.plot(np.sin(np.linspace(-np.pi/2, np.pi/2, 20)), 'cx:')
ax.plot(np.cos(np.linspace(-np.pi/2, np.pi/2, 20)), '--r8')
ax.legend(['Sine wave', 'Cosine wave'], loc='lower right', facecolor='coral', framealpha=0.4)
```

# Fill between

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y = np.sin(x)

plt.plot(x, y)
plt.fill_between(x, y, y2=0.25)
```

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y = np.sin(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(y > 0), color='g', alpha=.5)
plt.fill_between(x, y, where=(y < 0), color='r', alpha=.5)
```
