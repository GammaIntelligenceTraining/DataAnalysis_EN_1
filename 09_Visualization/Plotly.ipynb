```python
import plotly.express as px
```

```python
iris = px.data.iris()
gm = px.data.gapminder()
```

# Scatterplot

```python
iris.head()
```

```python
px.scatter(data_frame=iris,
           x='sepal_length',
           y='petal_length',
           height=450,
           color='species',
           hover_data=['sepal_width', 'petal_width'],
           marginal_y='violin',
           trendline='ols',
           color_discrete_map={'versicolor': 'orange', 'setosa': 'green', 'virginica': 'pink'}
          )
```

# Animated Scatterplot

```python
gm.head(5)
```

```python
px.scatter(gm,
           x='gdpPercap',
           y='lifeExp',
           height=600,
           color='continent',
           facet_col='continent',
           hover_name='country',
           size='pop',
           size_max=50,
           animation_frame='year',
           animation_group='country',
           range_y=[20, 100],
           log_x=True,
           title='Life expectation dynamics'
          )
```

# Bar plot

```python
px.bar(gm,
       x='year',
       y='pop',
       height=500,
       color='continent',
       barmode='group',
       hover_data=gm.columns
      )
```

# Funnel

```python
sales = {'stages': ['Awarness', 'Interest', 'Consideration', 'Intent', 'Evaluation', 'Purchase'],
         'values': [1000, 860, 590, 350, 120, 45]}

px.funnel(data_frame=sales,
          y='values',
          x='stages',
          height=500,
          orientation='v'
         )
```

# Pie plot

```python
gm.head()
```

```python
px.pie(data_frame=gm[gm.year == 2007],
       names='continent',
       values='pop',
       height=500,
       hole=.4
      )
```

# Sunburst

```python
px.sunburst(data_frame=gm[gm.year == 2007],
            path=['continent', 'country'],
            values='pop',
            height=600,
            color='lifeExp',
            color_continuous_scale=px.colors.diverging.curl
           )
```

# Treemap

```python
px.treemap(data_frame=gm[gm.year == 2007],
           path=[px.Constant('World'), 'continent', 'country'],
           values='pop',
           color='lifeExp',
           height=700,
           hover_data='gdpPercap'
          )
```

# Scatted 3d

```python
iris.head()
```

```python
px.scatter_3d(iris,
              x='sepal_length',
              y='sepal_width',
              z='petal_length',
              height=600,
              color='species'
             )
```

# Outline map

```python
gm.head()
```

```python
px.scatter_geo(gm,
               locations='iso_alpha',
               height=650,
               color='continent',
               size='pop',
               size_max=30,
               animation_frame='year',
               hover_name='country',
               projection='natural earth'
              )
```

# Tile map

```python
el = px.data.election()
el.head()
```

```python
gj = px.data.election_geojson()
```

```python
gj
```

```python
px.choropleth_mapbox(data_frame=el,
                     geojson=gj,
                     locations='district_id',
                     featureidkey='id',
                     mapbox_style='carto-positron',
                     height=600,
                     color='Bergeron',
                     center={'lat': 45.6, 'lon': -73.6},
                     zoom=9
                    )
```
