### Nested json example

```python
users = pd.read_json('Data/users.json')
```

```python
users.head()
```

```python
import json
```

```python
with open('Data/users.json') as file:
    users_dict = json.load(file)
```

```python
users_dict
```

```python
type(users_dict)
```

```python
users = pd.DataFrame.from_dict(users_dict['info'])
```

```python
users.head()
```

```python
users = pd.json_normalize(users_dict['info'], sep='_')
users.head()
```

```python
address = pd.json_normalize(users_dict, sep='_', record_path=['info', 'address'])
address.head()
```

```python
users.drop('address', axis=1, inplace=True)
```

```python
users.join(address)
```

```python
users = pd.concat([users.iloc[:, :4], address, users.iloc[:, 4:]], axis=1)
```

```python
users
```

### Save to json

```python
users.to_json()
```

```python
users.to_json('new_users.json')
```

```python
pd.read_json('new_users.json')
```

# Excel files

```python
df = pd.read_excel('Data/products.xlsx', index_col='product_id')
df.head()
```

```python
merchants = pd.read_excel('Data/products.xlsx', sheet_name='Merchants', index_col='merchant_id')
```

```python
merchants
```

### ExcelFile class

```python
excel_file = pd.ExcelFile('Data/products.xlsx')
```

```python
excel_file.sheet_names
```

```python
products = excel_file.parse('Products')
products.head()
```

```python
merchants = excel_file.parse('Merchants')
merchants.head()
```

```python
df = products.merge(merchants)
```

```python
df
```

```python
len(merchants)
```

## Save to excel

```python
df.to_excel('new_products.xlsx', index=None, sheet_name='Products')
```

### ExcelWriter

```python
writer = pd.ExcelWriter('new_excel_file.xlsx')
```

```python
with writer:
    products.to_excel(writer, sheet_name='Products', index=None)
    merchants.to_excel(writer, sheet_name='Merchants', index=None)
```
