```python
import json
```

```python
with open('Data/users.json') as file:
    users_dict = json.load(file)
```

```python
users_dict
```

```python
type(users_dict)
```

```python
users = pd.DataFrame.from_dict(users_dict['info'])
```

```python
users.head()
```

```python
users = pd.json_normalize(users_dict['info'], sep='_')
users.head()
```

```python
address = pd.json_normalize(users_dict, sep='_', record_path=['info', 'address'])
address.head()
```

```python
users.drop('address', axis=1, inplace=True)
```

```python
users.join(address)
```

```python
users = pd.concat([users.iloc[:, :4], address, users.iloc[:, 4:]], axis=1)
```

```python
users
```

# Request to API

```python
pd.Timestamp.now()
```

```python
str(int(pd.Timestamp(pd.Timestamp.now()).timestamp()))
```

```python
pd.to_datetime(1677024000, unit='s')
```

```python
def get_ohlc(fsym, tsym, limit=365, to_ts=pd.Timestamp.now(), all_data=False):
    url = 'https://min-api.cryptocompare.com/data/v2/histoday'

    to_ts = str(int(pd.Timestamp(to_ts).timestamp()))
    all_data = str(all_data).lower()

    params = {'fsym': fsym,
              'tsym': tsym,
              'limit': str(limit),
              'toTs': to_ts,
              'allData': all_data}

    response = requests.get(url, params=params)
    response.raise_for_status()
    data = response.json()

    df = pd.DataFrame.from_dict(data['Data']['Data']).drop(['volumefrom', 'conversionType', 'conversionSymbol'], axis=1)
    df.time = pd.to_datetime(df.time, unit='s')
    df.set_index('time', inplace=True)
    df.insert(0, 'open', df.pop('open'))
    df.insert(4, 'volume', df.pop('volumeto'))

    return df
```

```python
get_ohlc('BTC', 'USD', limit=90, to_ts='2023-01-01', all_data=True)
```

```python
pre_month = pd.Timestamp.now() - pd.offsets.Day(30)
```

```python
pre_month
```

```python
btc = get_ohlc('BTC', 'EUR', limit=30, to_ts=pre_month)
```

```python
eth = get_ohlc('ETH', 'EUR', limit=30, to_ts=pre_month)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
btc.close.plot(ax=ax1, title='Bitcoin')
eth.close.plot(ax=ax2, title='Etherium')
fig.tight_layout()
```
